```python
# https://www.bbc.co.uk/search?q=tesla&page=1

# http://feeds.bbci.co.uk/news/rss.xml


# https://www.healthdata.gov/

# <channel>
# <item>
# <description>

# get, post

import requests
from bs4 import BeautifulSoup

news_url = "http://feeds.bbci.co.uk/news/rss.xml"
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36'}

req = requests.get(news_url, headers = headers)
html = req.text

print(html)


# news_titles = soup.select('itembox > paditembox > item > ul > li > item >  div')
# news_desc = soup.select('itembox > paditembox > item > ul > li >  div')

```

```python
soup = BeautifulSoup(html, 'html.parser')

title_list = soup.find_all("title")

print(title_list)
```

```python
desc_list = soup.find_all("description")

print(desc_list)
```

```python
rows = []

items = soup.find_all("item")

for item in items:
    # print(item)
    title = item.find_all('title')[0].get_text()
    desc = item.find_all('description')[0].get_text()
    rows.append(f'{title} : {desc}')


for row in rows:
    print(row)
```
